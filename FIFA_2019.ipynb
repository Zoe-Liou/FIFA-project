# Data Cleaning

```python
## drop掉不需要的欄位（以難量化欄位為主）
fifa.drop(['ID', 'Name', 'Photo', 'Nationality', 'Flag', 'Club', 'Club Logo', 'Body Type', 'Real Face', 'Jersey Number', 'Joined', 'Loaned From', 'Contract Valid Until'],axis=1, inplace=True)
fifa.head()
```

```python
#查看缺失值
fifa.isnull()
```

```python
#drop含有任何NaN的row
fifa.dropna(axis= 0, how = 'any', inplace=True)
```

```python
#重新檢視缺失值
fifa.isnull().sum()
```

```python
#重新檢視資料
fifa
```

```python
#分別將含有價錢欄位的€符號去掉
fifa['Value'] = fifa['Value'].apply(lambda price : price.strip('€'))
fifa['Wage'] = fifa['Wage'].apply(lambda price : price.strip('€'))
fifa['Release Clause'] = fifa['Release Clause'].apply(lambda price : price.strip('€'))
```

```python
fifa['Wage'].unique()
```

```python
#將單位符號M和K轉換
def new_value(i):
    if 'M' in i:
        x = i[:-1]
        x = float(x)*1000000
        return(x)
    elif 'K' in i:
        x = i[:-1]
        x = float(x)*1000
        return(x)
    else:
        return None

fifa["Value"] = fifa["Value"].apply(new_value)
fifa["Wage"] = fifa["Wage"].apply(new_value)
fifa["Release Clause"] = fifa["Release Clause"].apply(new_value)
```

```python
#去除加號，並且以平均值代替
fifa['LS'].unique()
```

```python
def new_score(i):
    if '+' in i:
        x = i[:-2]
        x = int(x)+2
        return(x)
    else:
        return None
    
fifa["LS"] = fifa["LS"].apply(new_score)
fifa["ST"] = fifa["ST"].apply(new_score)
fifa["RS"] = fifa["RS"].apply(new_score)
fifa["LW"] = fifa["LW"].apply(new_score)
fifa["LF"] = fifa["LF"].apply(new_score)
fifa["CF"] = fifa["CF"].apply(new_score)
fifa["RF"] = fifa["RF"].apply(new_score)
fifa["RW"] = fifa["RW"].apply(new_score)
fifa["LAM"] = fifa["LAM"].apply(new_score)
fifa["CAM"] = fifa["CAM"].apply(new_score)
fifa["RAM"] = fifa["RAM"].apply(new_score)
fifa["LM"] = fifa["LM"].apply(new_score)
fifa["LCM"] = fifa["LCM"].apply(new_score)
fifa["CM"] = fifa["CM"].apply(new_score)
fifa["RCM"] = fifa["RCM"].apply(new_score)
fifa["RM"] = fifa["RM"].apply(new_score)
fifa["LWB"] = fifa["LWB"].apply(new_score)
fifa["LDM"] = fifa["LDM"].apply(new_score)
fifa["CDM"] = fifa["CDM"].apply(new_score)
fifa["RDM"] = fifa["RDM"].apply(new_score)
fifa["RWB"] = fifa["RWB"].apply(new_score)
fifa["LB"] = fifa["LB"].apply(new_score)
fifa["LCB"] = fifa["LCB"].apply(new_score)
fifa["CB"] = fifa["CB"].apply(new_score)
fifa["RCB"] = fifa["RCB"].apply(new_score)
fifa["RB"] = fifa["RB"].apply(new_score)
```

```python
#去掉體重單位轉乘kg
fifa["Weight"] = fifa["Weight"].str.strip("lbs")
fifa["Weight"]=pd.DataFrame(fifa.Weight,dtype=np.float)
fifa["Weight"] = fifa["Weight"]*0.45
fifa["Weight"]=pd.DataFrame(fifa.Weight,dtype=np.float)
```

```python
#英呎去除引號並換成公分
fifa["Height"] = fifa["Height"].str.replace("'","")
fifa["Height"]=pd.DataFrame(fifa.Height,dtype=np.int)
```

```python
fifa["Height"].unique()
```

```python
def new_height(x):
    if x<=100:
        y = x/10
        return(y)
    elif x >=100:
        y = x/100
        return(y)
    else:
        return None
fifa["Height"] = fifa["Height"].apply(new_height)
```

```python
fifa["Height"].unique()
```

```python
fifa["Height"] = (fifa["Height"]*30.48)/100
```

```python
#確定身高英呎已轉換成公尺
fifa["Height"].unique()
```

```python
#身高體重換成BMI
fifa['BMI'] = fifa['Weight']/fifa['Height']**2
```

```python
#將慣用腳轉換成1和0
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

fifa['Preferred Foot'] = label_encoder.fit_transform(fifa['Preferred Foot'])
fifa['Work Rate'] = label_encoder.fit_transform(fifa['Work Rate'])
```

```python
#利用該位置積分最高為主，各位置的評分轉化成最擅長位置評分
positions = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']

fifa['best_pos'] = fifa[positions].max(axis =1)
```

```python
fifa.drop(['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB'],axis=1, inplace=True)
fifa.head()
```

# Decision Tree Regression

```python
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn.externals import joblib
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
fifa1.drop(['Position'],axis=1, inplace=True) #drop掉非數字格式的“位置”來進行決策樹迴歸分析
fifa1.head()
```

```python
fifa1.info()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
#拆分訓練與測試data
X_train, X_test, y_train, y_test = train_test_split(fifa1.drop('Release Clause',axis=1), fifa1['Release Clause'], test_size=0.30, random_state=100)
```

```python
#選擇六個特徵決策樹標竿模型
col = ['Potential', 'International Reputation', 'Age', 'best_pos', 'Reactions', 'Value']
```

```python
reg = DecisionTreeRegressor(random_state=100)
```

```python
reg.fit(X_train[col], y_train.astype('int'))
```

```python
pre = reg.predict(X_test[col])
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print('R-square: %.2f' % r2_score(y_test,pre))
```

```python
plt.scatter(y_test,pre)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
sns.distplot((y_test-pre),bins=40);
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, pre))
print('MSE:', metrics.mean_squared_error(y_test, pre))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pre)))
```
